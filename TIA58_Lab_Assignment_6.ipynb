```python
import csv
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori , association_rules
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
X = pd.read_csv('/Coursetopics.csv')
```

```python
X
```

```python
X_bool = X.astype(bool)
X_bool
```

```python
frequent_items= apriori(X_bool,min_support=2/365,use_colnames=True)
```

```python
frequent_items
```

```python
rules = association_rules(frequent_items,metric = 'confidence',min_threshold=183/365)
```

```python
rules
```

```python
rules=rules[['antecedents','consequents','support','confidence']]
```
